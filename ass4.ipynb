```python
import pandas as pd
# Load dataset
data = pd.read_csv("multiple_linear_regression_dataset.csv")
# Inspect data
print(data.head())
print(data.columns)
print(data.shape)
```

```python
# input features are age and experience and output feature is income and model needs to handle 2 features
# Inputs (features)
X = data[["experience","age"]]
# Output (target)
y = data["income"]
```

```python
import numpy as np
# Number of features
n_features = X.shape[1]
# Initialize weights and bias
w = np.zeros(n_features)
b =0.0
```

```python
# we need one weight per feature beacuse each feature influences the prediction differently, so the model needs a separate parameter to measure its importance.
# bias allows the model to make predictions even when all inputs are zero.
# yes, because of sigmoid saturation , unstable updates and poor learning.
```

```python
def predict(X, w, b):
    """
    X : input features
    w : weights
    b : bias
    returns predicted values
    """
    y_hat = X.dot(w) + b
    return y_hat
```

```python
# No activation is needed because: Outputs are not probabilities, We do not need values restricted between 0 and 1, A linear relationship is sufficient
# y_hat can take any real number
# Logistic regression adds a sigmoid activation to convert outputs into probabilities for classification.
```

```python
def mean_squared_error(y, y_hat):
    """
    y     : actual values
    y_hat : predicted values
    """
    loss = ((y_hat - y) ** 2).mean()
    return loss
```

```python
# We square errors to avoid cancellation and penalize large mistakes more strongly.
# Because the error is squared, large mistakes become much larger which forces model to correct big mistakes
# Absolute error is less preferred because it is not smoothly differentiable and gives less stable gradients.
```

```python
# compute gradients
# without gradients, weights never change and loss never improves
def compute_gradients(X, y, y_hat):
    """
    Compute gradients of MSE loss
    """
    N = len(y)

    # Gradient w.r.t weights
    dw = (2 / N) * X.T.dot(y_hat - y)

    # Gradient w.r.t bias
    db = (2 / N) * (y_hat - y).sum()

    return dw, db
```

```python
# Because weights multiply the input features, but bias does not.
# Error appears everywhere because gradients measure how prediction mistakes change parameters.
# If error = 0, gradients become zero and learning stops.
```

```python
# update parameters
def update_parameters(w, b, dw, db, lr):
    w = w - lr * dw
    b = b - lr * db
    return w, b
```

```python
# training loop
lr = 0.0001
epochs = 1000

for epoch in range(epochs):
    y_hat = predict(X, w, b)
    loss = mean_squared_error(y, y_hat)
    dw, db = compute_gradients(X, y, y_hat)
    w, b = update_parameters(w, b, dw, db, lr)

    if epoch % 100 == 0:
        print(f"Epoch {epoch}, Loss:{loss}")
```

```python
# Loss should decrease as the model learns better parameters.
# If loss increases, learning rate may be too high or gradients incorrect.(Unstable Learning)
# Learning rate controls step size while epochs control how long training runs; they must be balanced.
```

```python
# final evaluation
new_candidate = np.array([4.5, 68])
predicted_salary = new_candidate.dot(w) + b
print(predicted_salary)
```

```python
# yes, prediction is reasonable if the input lies within the training distribution
# linear regression interpolates smoothly because outputs change continuously with inputs (small input change equals small output change)
# it is better than threshold rules because it learns weighted relationships from data instead of using rigid conditions
```
