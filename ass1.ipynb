```python
class Perceptron:
    def __init__(self, lr=0.1, epochs=10):
        self.lr = lr
        self.epochs = epochs
        self.weights = [0, 0]   # w1, w2
        self.bias = 0

    def activation(self, net):
        return 1 if net >= 0 else 0

    def predict(self, x):
        net = x[0] * self.weights[0] + x[1] * self.weights[1] + self.bias
        return self.activation(net)

    def train(self, X, y):
        for _ in range(self.epochs):
            for i in range(len(X)):
                y_pred = self.predict(X[i])
                error = y[i] - y_pred

                self.weights[0] += self.lr * error * X[i][0]
                self.weights[1] += self.lr * error * X[i][1]
                self.bias += self.lr * error

X = [
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
]

# AND Gate

print("\nAND Gate")
y_and = [0, 0, 0, 1]

p = Perceptron()
p.train(X, y_and)

print("Weights:", p.weights)
print("Bias:", p.bias)
for x in X:
    print(x, "->", p.predict(x))


# OR Gate

print("\nOR Gate")
y_or = [0, 1, 1, 1]

p = Perceptron()
p.train(X, y_or)

print("Weights:", p.weights)
print("Bias:", p.bias)
for x in X:
    print(x, "->", p.predict(x))


# NAND Gate

print("\nNAND Gate")
y_nand = [1, 1, 1, 0]

p = Perceptron()
p.train(X, y_nand)

print("Weights:", p.weights)
print("Bias:", p.bias)
for x in X:
    print(x, "->", p.predict(x))


# NOR Gate

print("\nNOR Gate")
y_nor = [1, 0, 0, 0]

p = Perceptron()
p.train(X, y_nor)

print("Weights:", p.weights)
print("Bias:", p.bias)
for x in X:
    print(x, "->", p.predict(x))


# XOR Gate (Will Fail)

print("\nXOR Gate")
y_xor = [0, 1, 1, 0]

p = Perceptron(epochs=50)
p.train(X, y_xor)

print("Weights:", p.weights)
print("Bias:", p.bias)
for x in X:
    print(x, "->", p.predict(x))
```

```python
# Learning rate controls how slow or how fast weight should get updated, 
# in case of linearly separable gates learning rate only convergence speed and not final result
# same code learned different gates because perceptron equation remains same, only weights are adjusted
# XOR cannot be learned because it is not linearly separable
```
